```python
# Drop specified columns from the DataFrame
data.drop(['infant deaths','percentage expenditure','GDP'], axis=1, inplace=True)
```

```python
#  label encoder  data to convert object to int
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in data.select_dtypes(include='object'):
    data[col] = le.fit_transform(data[col])
```

```python
# remove outliers by IQR

Q1 = data['Life expectancy'].quantile(0.25)
Q3 = data['Life expectancy'].quantile(0.75)
IQR = Q3 - Q1

```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

data = data[(data['Life expectancy'] >= lower_bound) & (data['Life expectancy'] <= upper_bound)]
```

```python
# replace null values in data

data.fillna(data.mean(), inplace=True)
```

APPLY  ALGORITHM TO DATASET

Linear rigressor

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


x2=data.drop('Life expectancy', axis=1)
y2 = data['Life expectancy']
```

```python
print(x2)
```

```python
from sklearn.preprocessing import StandardScaler #arrange the data based on range (-3 to +3)
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split  #train and test

x_train,x_test,y_train,y_test=train_test_split(x2,y2,test_size=0.2,random_state=40)
#standardize the features using StandardScaler

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train) #Fit-used to fit, transform->change values 1 to another
x_test=scaler.transform(x_test)

#Train the linear regression model

lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error

y_pred = lin_reg.predict(x_test)

print("Mean squared error:", mean_squared_error(y_test, y_pred))
print("R-squared score:", r2_score(y_test, y_pred))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

# Create a decision tree regressor
decision_tree = DecisionTreeRegressor()

# Train the decision tree regressor
decision_tree.fit(x_train, y_train)

# Predict the life expectancy using the decision tree
y_pred_dt = decision_tree.predict(x_test)

# Evaluate the decision tree regressor
print("Mean squared error:", mean_squared_error(y_test, y_pred_dt))
print("R-squared score:", r2_score(y_test, y_pred_dt))
```

Random Forest

```python
# Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor

# Create a random forest regressor
random_forest = RandomForestRegressor()

# Train the random forest regressor
random_forest.fit(x_train, y_train)

# Predict the life expectancy using the random forest
y_pred_rf = random_forest.predict(x_test)

# Evaluate the random forest regressor
print("Mean squared error:", mean_squared_error(y_test, y_pred_rf))
print("R-squared score:", r2_score(y_test, y_pred_rf))
```

```python
# K-Nearest Neighbors

from sklearn.neighbors import KNeighborsRegressor

# Create a KNN regressor
knn = KNeighborsRegressor()

# Train the KNN regressor
knn.fit(x_train, y_train)

# Predict the life expectancy using the KNN regressor
y_pred_knn = knn.predict(x_test)

# Evaluate the KNN regressor
print("Mean squared error:", mean_squared_error(y_test, y_pred_knn))
print("R-squared score:", r2_score(y_test, y_pred_knn))
```

Support Vector Regressor

```python
from sklearn.svm import SVR

# Create a support vector regression model
svm_regressor = SVR()

# Train the support vector regression model
svm_regressor.fit(x_train, y_train)

# Predict the life expectancy using the support vector regression model
y_pred_svm = svm_regressor.predict(x_test)

# Evaluate the support vector regression model
print("Mean squared error:", mean_squared_error(y_test, y_pred_svm))
print("R-squared score:", r2_score(y_test, y_pred_svm))
```

```python
# based on the accuracy compare each model accuracy in table and highlight high accuracy

models = ['Linear Regression', 'Decision Tree', 'Random Forest', 'KNN', 'SVM']
mse = [mean_squared_error(y_test, y_pred), mean_squared_error(y_test, y_pred_dt), mean_squared_error(y_test, y_pred_rf), mean_squared_error(y_test, y_pred_knn), mean_squared_error(y_test, y_pred_svm)]
r2 = [r2_score(y_test, y_pred), r2_score(y_test, y_pred_dt), r2_score(y_test, y_pred_rf), r2_score(y_test, y_pred_knn), r2_score(y_test, y_pred_svm)]

results = pd.DataFrame({'Model': models, 'MSE': mse, 'R2 Score': r2})

# Highlight the highest accuracy in green
results.style.apply(lambda x: ['background-color: green' if x.name == results['R2 Score'].idxmax() else '' for i in x], axis=1)
```
